# AMEO - Analysis of Aspiring Mind Employment Outcome 2015 (AMEO)

## Dataset information : 

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate.

## Objective :

With salary serving as the dependent variable, the aim of this Exploratory Data Analysis (EDA) is to extensively investigate the provided dataset, focusing particularly on understanding the relationships between various variables and the target variable, Salary.

###### The columns 'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', and 'CivilEngg' represent scores from the AMCAT test. Having -1 values in these columns seems inappropriate. We'll explore if there are outliers present, based on which we can decide whether to replace the -1 values with the median or mean.

```python
columns = ['Domain', 'ComputerProgramming', 'ElectronicsAndSemicon', 
           'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 
           'TelecomEngg', 'CivilEngg']
def detect_outliers_iqr(data):
    outliers = {}
    for column in columns:
        Q1 = np.percentile(data[column], 25)
        Q3 = np.percentile(data[column], 75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_indices = np.where((data[column] < lower_bound) | (data[column] > upper_bound))
        if len(outlier_indices[0]) > 0:
            outliers[column] = outlier_indices[0]
    return outliers
outliers = detect_outliers_iqr(df)
if outliers:
    print("Columns with outliers:")
    for column, indices in outliers.items():
        print(f"{column}: {indices}")
else:
    print("No outliers found in the specified columns.")
```

```python
#since all columns have outliers replacing it with median 
```

```python
Columns = ['Domain','ComputerProgramming','ElectronicsAndSemicon','ComputerScience',
          'MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']
for i in Columns:
    median = df[i][df[i] != -1].median()
    df[i] = df[i].replace(-1, median)
```
